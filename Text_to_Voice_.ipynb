```python
from gtts import gTTS

text_to_say = " This is a sample piece of text read by GTTS."

language = "en"

gtts_object = gTTS(text = text_to_say,
                  lang = language,
                  slow = False)

gtts_object.save("/content/gtts.wav")
```

```python
french_text = "Bonjour. Je m'appelle Alexandra. Je suis programmeuse."

french_language = "fr"

french_gtts_object = gTTS(text = french_text,
                          lang = french_language,
                          slow = True)

french_gtts_object.save("/content/french.wav")

Audio("/content/french.wav")
```

```python
# Define the Bangla text and language
bangla_text = "কেমন আছেন আপনি, আশা করি ভাল আছেন। আমার এই প্রজেক্ট কেমন লেগেছে আপনার কাছেন জানাবেন"
bangla_language = "bn"

# Create gTTS object
bangla_gtts_object = gTTS(text=bangla_text, lang=bangla_language, slow=False)

# Save the audio file as MP3
output_file = "/content/bangla.mp3"
bangla_gtts_object.save(output_file)

# Play the audio file in a Jupyter notebook or similar environment
Audio(output_file)
```

```python
# for additional speed change
! pip install pydub
from pydub import AudioSegment
from pydub.playback import play

```

```python
# Load the MP3 file
audio = AudioSegment.from_mp3(output_file)

# Function to change speed
def change_speed(audio_segment, speed_change):
    # Calculate new sample rate
    new_sample_rate = int(audio_segment.frame_rate * speed_change)
    # Change sample rate of the audio segment
    return audio_segment._spawn(audio_segment.raw_data, overrides={'frame_rate': new_sample_rate}).set_frame_rate(audio_segment.frame_rate)

# Speed up the audio by 1.5 times (faster)
fast_audio = change_speed(audio, 1.2)
fast_audio_file = "/content/bangla_fast.mp3"
fast_audio.export(fast_audio_file, format="mp3")

# Slow down the audio by 0.5 times (slower)
slow_audio = change_speed(audio, 0.9)
slow_audio_file = "/content/bangla_slow.mp3"
slow_audio.export(slow_audio_file, format="mp3")

# Play the modified audio files in a Jupyter notebook or similar environment
from IPython.display import display
display(Audio(fast_audio_file))
display(Audio(slow_audio_file))
```
